### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
# Get the total rides for each city type
total_rides = pyber_data_df.groupby('type').count()['ride_id']

total_rides
```

```python
# Get the total drivers for each city type
total_drivers = (pyber_data_df.drop_duplicates(['city'])
                     .groupby('type')
                     .sum(numeric_only=True)['driver_count'])

total_drivers
```

```python
# Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby('type').sum(numeric_only=True)['fare']

total_fares
```

```python
# Get the average fare per ride for each city type. 
avg_fare_per_ride = total_fares / total_rides

avg_fare_per_ride
```

```python
# Get the average fare per driver for each city type. 
avg_fare_per_driver = total_fares / total_drivers

avg_fare_per_driver
```

```python
# Create a PyBer summary DataFrame.
summary_data = {
    "Total Rides" : total_rides,
    "Total Drivers" : total_drivers,
    "Total Fares" : total_fares,
    "Average Fare per Ride" : avg_fare_per_ride,
    "Average Fare per Driver" : avg_fare_per_driver
    }
pyber_summary_df = pd.DataFrame(summary_data)

pyber_summary_df.head()
```

```python
# Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
# Format the columns.
pyber_summary_df.style.format({'Total Rides':'{:,}',
                               'Total Drivers':'{:,}',
                               'Total Fares': '${:,.2f}',
                               'Average Fare per Ride': '${:,.2f}',
                               'Average Fare per Driver': '${:,.2f}'})
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Read the merged DataFrame
pyber_data_df.head()
```

```python
# Using groupby() to create a new DataFrame showing the sum of the fares 
# for each date where the indices are the city type and date.
fare_by_type_date = pyber_data_df.groupby(['type', 'date']).sum(numeric_only=True)['fare']

fare_by_type_date = pd.DataFrame(fare_by_type_date)
fare_by_type_date
```

```python
# Reset the index on the DataFrame. This is needed to use the 'pivot()' function.
fare_by_type_date = fare_by_type_date.reset_index()

fare_by_type_date
```

```python
# Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_pivot = pd.pivot(fare_by_type_date,
                      index='date', columns='type', values='fare')

fare_pivot
```

```python
# Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_to_april_fares = fare_pivot.loc['2019-01-01':'2019-04-29']

jan_to_april_fares
```

```python
# Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
jan_to_april_fares.index = pd.to_datetime(jan_to_april_fares.index)
```

```python
# Check that the datatype for the index is datetime using df.info()
jan_to_april_fares.info()
```

```python
# Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_by_week = jan_to_april_fares.resample('W').sum()

fares_by_week
```

```python
# Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.figure()
weekly_fares_plot = fares_by_week.plot(figsize=(14, 4))
weekly_fares_plot.set_title('Total Fares by City Type')
weekly_fares_plot.set_ylabel('Fares ($USD)')
weekly_fares_plot.set_xlabel('')
weekly_fares_plot.legend(loc=(0.5, 0.5))

plt.savefig('Analysis/PyBer_fare_summary.png')

plt.show()
```
